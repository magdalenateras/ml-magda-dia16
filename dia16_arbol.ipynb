```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=200, n_features=6, n_informative=4, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X_train, y_train)
pred = tree.predict(X_test)
print(f"Accuracy Tree: {accuracy_score(y_test, pred):01%}")
```

```python
features = ['horas_trabajo', 'sueno', 'estres', 'edad', 'cafe_diario', 'ejercicio']
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,12))
plot_tree(tree, feature_names=features, class_names=['No_Burnout', 'Burnout'], filled=True, rounded=True, fontsize=10)
plt.savefig('dia16_arbol.png', dpi=150, bbox_inches='tight')
plt.show()
print("dia16_arbol.png CREADO")
```

```python
importances= {features[i]: f"{tree.feature_importances_[i]:.2f}" for i in range(6)}
print("Importances:", importances)
```

```python
#Top features ordenadas
importances_ordenadas = sorted(zip(features, tree.feature_importances_),
                                key=lambda x: x[1], reverse=True)
print("\n FEATURES IN ORDER:")
for i, (feature, imp) in enumerate(importances_ordenadas, 1):
    print(f"{i}. {feature}: {imp:02%}")

#grafico
plt.figure(figsize=(10,6))
features_top = [f[0] for f in importances_ordenadas]
imp_top = [f[1] for f in importances_ordenadas]
plt.barh(features_top, imp_top, color='gold')
plt.xlabel('Importances')
plt.title('Dia 16: Features Burnout Predictor ORDENADAS')
plt.tight_layout()
plt.savefig('dia16_importances.png', dpi=150)
plt.show()
print("dia16_importances.png CREADO!")
print("Dia 16 COMPLETADO!")
```
